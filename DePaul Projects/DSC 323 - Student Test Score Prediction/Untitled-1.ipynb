```python
def get_measures(data, latex=False):
    mean = np.mean(data)
    median = np.median(data)
    mode = data.mode().values[0]
    
    std = np.std(data)
    var = np.var(data)
    max = np.max(data)
    min = np.min(data)
    range = max - min
    iqr = stats.iqr(data)
    skew = stats.skew(data)
    
    df = pd.DataFrame({
        'Values': [mean, median, mode, std, var, min, max, range, iqr, skew]
    }, index=['Mean', 'Median', 'Mode', 'St. Dev.', 'Variance', 'Min', 'Max', 'Range', 'IQR', 'Skew'])
    df.index.name = 'Measure'
    if latex:
        return print(df.to_latex().replace('tabular', 'tabu'))
    else:
        return df
    
    
```
